## データの読み込み

推計値であることに注意

```python
# UserWarningが消せないため、　Unknown type for savedVersions　のみ無視する
import warnings
warnings.filterwarnings("ignore", category=UserWarning, message="Unknown type for SavedVersions")


df_origin = pd.read_excel("../data/2023_広島県_理由別移動者数.xlsx", index_col=0, engine='openpyxl')
```

```python
hiroshima_headers = ["総計", 
           "移動の主因者_総数", "移動の主因者_就職", "移動の主因者_転勤", "移動の主因者_転業・転職", "移動の主因者_退職・廃業", 
           "移動の主因者_入学・転校", "移動の主因者_通勤・通学の便",
           "移動の主因者_結婚・離婚・養子縁組", "移動の主因者_子育て環境上", "移動の主因者_介護", "移動の主因者_住宅事情", 
           "移動の主因者_その他", "移動の主因者_不詳", 
           "主因者に伴う者"]

# 広島県の転入・転出の推計値
df_moving_in_2023 = df_origin[6:20]
df_moving_in_2023 = df_moving_in_2023.set_axis(hiroshima_headers, axis=1).astype(float)

df_moving_out_2023 = df_origin[25:39]
df_moving_out_2023 = df_moving_out_2023.set_axis(hiroshima_headers, axis=1).astype(float)

df_moving_in_2023.to_csv("../data/out/moving_in_2023.csv")
df_moving_out_2023.to_csv("../data/out/moving_out_2023.csv")
```

```python
tokushima_headers = ["総数", "徳島市", "鳴門市", "小松島市", "阿南市", "吉野川市", "阿波市", "美馬市", "三好市", "勝浦町", "上勝町", "佐那河内村", "石井町", "神山町", "那賀町", "牟岐町",
                     "美波町", "海陽町", "松茂町", "北島町", "藍住町", "板野町", "上板町", "つるぎ町", "東みよし町"]

df_tokushima = pd.read_excel("../data/2023_徳島県_移動状況表.xlsx", index_col=0, engine='openpyxl', sheet_name="5歳階級別　県内市町村間の移動状況")

df_tokushima_moving_in_2023 = df_tokushima[6:28]
df_tokushima_moving_in_2023 = df_tokushima_moving_in_2023.set_axis(tokushima_headers, axis=1).astype(int)
df_tokushima_moving_out_2023 = df_tokushima[37:59]
df_tokushima_moving_out_2023 = df_tokushima_moving_out_2023.set_axis(tokushima_headers, axis=1).astype(int)

df_tokushima_moving_in_2023["神山町"].to_csv("../data/out/tokushima_moving_in_2023.csv")
df_tokushima_moving_out_2023["神山町"].to_csv("../data/out/tokushima_moving_out_2023.csv")
```

```python
index_col = ["年度", "調査種別", "地域", "データ種別", "因子群", "カテゴリ・評価指標"]

df_well_being_hiroshima_2023 = pd.read_csv("../data/2023年度版(令和5年度版) Well-Being 全国調査_広島県全域（都道府県表示）_客観_utf-8.csv", index_col=index_col)
df_well_being_kamiyama_2023 = pd.read_csv("../data/2023年度版(令和5年度版) Well-Being 全国調査_徳島県神山町_客観_utf-8.csv", index_col=index_col)
df_well_being_hiroshima_2023.to_csv("../data/out/well_being_hiroshima_2023.csv")
df_well_being_kamiyama_2023.to_csv("../data/out/well_being_kamiyama_2023.csv")
```

```python
header = ["", "和暦", "西暦", 
          "全国_15歳以上人口", "全国_労働力人口総数", "全国_労働力人口就業者", "全国_労働力人口完全失業者",
          "全国_非労働力人口", "全国_労働力人口比率", "全国_完全失業率",
          "広島県_15歳以上人口", "広島県_労働力人口総数", "広島県_労働力人口就業者", "広島県_労働力人口完全失業者",
          "広島県_非労働力人口", "広島県_労働力人口比率", "広島県_完全失業率"]
labor_force_servey = pd.read_excel("../data/労働力基本調査.xlsx", usecols=range(2, 17), names=header, sheet_name="年平均（全国・広島県）")[5:32]

labor_force_servey = labor_force_servey.astype(
    {
        "西暦": np.int64,
        "全国_15歳以上人口": np.int64,
        "全国_労働力人口総数": np.int64,
        "全国_労働力人口就業者": np.int64,
        "全国_労働力人口完全失業者": np.int64,
        "全国_非労働力人口": np.int64,
        "全国_労働力人口比率": np.float64,
        "全国_完全失業率": np.float64,
        "広島県_15歳以上人口": np.int64,
        "広島県_労働力人口総数": np.int64,
        "広島県_労働力人口就業者": np.int64,
        "広島県_労働力人口完全失業者": np.int64,
        "広島県_非労働力人口": np.int64,
        "広島県_労働力人口比率": np.float64,
        "広島県_完全失業率": np.float64
    }
)
labor_force_servey = labor_force_servey.astype({"西暦": object})

labor_force_servey.to_csv("../data/out/labor_force_servey.csv")
```
