### 人口流入数が増加している自治体
1. 神山町（徳島県） - 都市部からの移住者を惹きつける政策を進め、特にテレワークの推進や企業の誘致に成功しています。
1. 嬉野市（佐賀県） - 温泉地としての魅力を活かし、移住支援策を充実させ人口流入を促進しています。
1. 小布施町（長野県） - 豊かな自然と芸術文化が魅力で、若者やアーティストの移住が増えています。
1. 南城市（沖縄県） - 沖縄の自然と文化を生かした移住政策が功を奏し、多くの人々が移住しています。
1. 上天草市（熊本県） - 海と自然に囲まれ、地域全体で移住者を受け入れる体制を整えています。
1. 野沢温泉村（長野県） - 温泉とウィンタースポーツが魅力の地域で、国内外からの観光客に加えて移住者も増えています。
1. 大潟村（秋田県） - 農業が盛んで、若者の農業参入を支援する政策が功を奏しています。

```python
df_well_being_hiroshima = pd.read_csv('../data/2023年度版(令和5年度版) Well-Being 全国調査_広島県全域（都道府県表示）_客観.csv', encoding='sjis')

df_well_being_kamiyama = pd.read_csv('../data/2023年度版(令和5年度版) Well-Being 全国調査_徳島県神山町_客観.csv', encoding='sjis')
df_well_being_ureshino = pd.read_csv("../data/2023年度版(令和5年度版) Well-Being 全国調査_佐賀県嬉野市_客観.csv", encoding='sjis')
df_well_being_obuse = pd.read_csv("../data/2023年度版(令和5年度版) Well-Being 全国調査_長野県小布施町_客観.csv", encoding='sjis')
df_well_being_nanjo = pd.read_csv("../data/2023年度版(令和5年度版) Well-Being 全国調査_沖縄県南城市_客観.csv", encoding='sjis')
df_well_being_kamiamakusa = pd.read_csv("../data/2023年度版(令和5年度版) Well-Being 全国調査_熊本県上天草市_客観.csv", encoding='sjis')
df_well_being_nozawaonsen = pd.read_csv("../data/2023年度版(令和5年度版) Well-Being 全国調査_長野県野沢温泉村_客観.csv", encoding='sjis')
df_well_being_ogata = pd.read_csv("../data/2023年度版(令和5年度版) Well-Being 全国調査_秋田県大潟村_客観.csv", encoding='sjis')
# データの先頭5行を表示
df_well_being_hiroshima = df_well_being_hiroshima.T[5:6]
df_well_being_hiroshima["自治体"] = "広島県"
df_well_being_hiroshima.set_index("自治体", inplace=True)

df_well_being_kamiyama = df_well_being_kamiyama.T[5:6]
df_well_being_kamiyama["自治体"] = "徳島県神山町"
df_well_being_kamiyama.set_index("自治体", inplace=True)

df_well_being_ureshino = df_well_being_ureshino.T[5:6]
df_well_being_ureshino["自治体"] = "佐賀県嬉野市"
df_well_being_ureshino.set_index("自治体", inplace=True)

df_well_being_obuse = df_well_being_obuse.T[5:6]
df_well_being_obuse["自治体"] = "長野県小布施町"
df_well_being_obuse.set_index("自治体", inplace=True)

df_well_being_nanjo = df_well_being_nanjo.T[5:6]
df_well_being_nanjo["自治体"] = "沖縄県南城市"
df_well_being_nanjo.set_index("自治体", inplace=True)

df_well_being_kamiamakusa = df_well_being_kamiamakusa.T[5:6]
df_well_being_kamiamakusa["自治体"] = "熊本県上天草市"
df_well_being_kamiamakusa.set_index("自治体", inplace=True)

df_well_being_nozawaonsen = df_well_being_nozawaonsen.T[5:6]
df_well_being_nozawaonsen["自治体"] = "長野県野沢温泉村"
df_well_being_nozawaonsen.set_index("自治体", inplace=True)

df_well_being_ogata = df_well_being_ogata.T[5:6]
df_well_being_ogata["自治体"] = "秋田県大潟村"
df_well_being_ogata.set_index("自治体", inplace=True)
```

## 移動人口
流入数と流出数をDataFrameに格納する

```python
# 広島県
df_hiroshima_moving_in = pd.read_csv("../data/out/moving_in_2023.csv")
df_hiroshima_moving_out = pd.read_csv("../data/out/moving_out_2023.csv")
df_hiroshima_moving_index = df_hiroshima_moving_in["Unnamed: 0"]
df_hiroshima_moving = pd.concat(
    [
        df_hiroshima_moving_index, 
        df_hiroshima_moving_in.drop(["Unnamed: 0"], axis=1) - df_hiroshima_moving_out.drop(["Unnamed: 0"], axis=1),
    ], axis=1)
df_hiroshima_moving.rename(columns={"Unnamed: 0": "5歳階級別"}, inplace=True)
df_hiroshima_moving
```

```python
# 徳島県神山町
df_kamiyama_moving_in = pd.read_csv("../data/out/tokushima_moving_in_2023.csv")
df_kamiyama_moving_out = pd.read_csv("../data/out/tokushima_moving_out_2023.csv")
df_kamiyama_moving_in.rename(columns={"5歳階級別\u3000県内市町村間の移動状況": "5歳階級別", "神山町": "総計"}, inplace=True)
df_kamiyama_moving_out.rename(columns={"5歳階級別\u3000県内市町村間の移動状況": "5歳階級別", "神山町": "総計"}, inplace=True)
df_kamiyama_moving_index = df_kamiyama_moving_in["5歳階級別"]
df_kamiyama_moving = pd.concat(
    [
        df_kamiyama_moving_index,
        df_kamiyama_moving_in.drop(["5歳階級別"], axis=1) - df_kamiyama_moving_out.drop(["5歳階級別"], axis=1)
    ], axis=1
)
over_65_sum = df_kamiyama_moving[
    df_kamiyama_moving["5歳階級別"].str.startswith(
        ("65", "70", "75", "80", "85", "90", "95", "100", "年齢不詳")
    )]["総計"].sum()
df_kamiyama_moving["5歳階級別"] = df_kamiyama_moving["5歳階級別"].apply(lambda x: f"{x}歳")
df_kamiyama_moving = pd.concat([df_kamiyama_moving[0:13], pd.DataFrame({"5歳階級別": "65歳以上", "総計": over_65_sum}, index=[0])], ignore_index=True)
df_kamiyama_moving.tail()
```

```python
# 佐賀県嬉野市
df_ureshino_moving = pd.read_csv("../data/out/ureshino_moving.csv")
df_ureshino_moving = df_ureshino_moving.drop(["Unnamed: 0", "西暦", "総計", "五町田", "久間", "塩田", "吉田"], axis=1)
df_ureshino_moving
```

```python
# 長野県小布施町
df_nagano = pd.read_excel("../data/2023_長野県_移動状況.xlsx", header=None)
df_obuse = df_nagano[df_nagano[1] == "小布施町"]
obuse_jan = df_obuse.iloc[0, [2, 3, 4, 5]].values
obuse_dec = df_obuse.iloc[0, [46, 47, 48, 49]].values
difference = obuse_dec - obuse_jan

header = ["世帯数", "総計", "男", "女"]
df_obuse_moving = pd.DataFrame([difference], columns=header)
df_obuse_moving
```

```python
# 沖縄県南城市
header=["市町村名", "2002年総計", "出生", "死亡", "自然増減", "自然増減率", "転入_県外", "転入_県内", "転入_その他", "転入_計",
        "転出_県外", "転出_県内", "転出_その他", "転出_計", "社会増減", "社会増減率", "人口増減数", "人口増減率", "2003年総計", "市町村名_別"]
df_nanjo_moving = pd.read_excel("../data/沖縄県_市町村別人口増減数.xls")[20:21]
df_nanjo_moving.columns = header
df_nanjo_moving.drop(columns=["2002年総計", "市町村名_別"], inplace=True)
df_nanjo_moving.reset_index(drop=True, inplace=True)
df_nanjo_moving
```

```python
# 熊本県上天草市
header = ["市町村名", "世帯数", "総人口", "男性", "女性", "人口性比（女=100）", "1世帯当たり人員",
          "純増減", "出生者", "死亡者", "自然増減", 
          "転入者_総数", "転入者_県内", "転入者_県外", "転入者_その他", "転出者_総数", "転出者_県内", "転出者_県外", "転出者_その他",
          "社会増減", "人口増減率"]
df_kamiamakusa_moving = pd.read_excel("../data/2023_熊本県_移動状況.xlsx")[22:23].drop(["Unnamed: 0", "Unnamed: 22", "Unnamed: 23", "Unnamed: 24"], axis=1)
df_kamiamakusa_moving.columns = header
df_kamiamakusa_moving.reset_index(drop=True, inplace=True)
df_kamiamakusa_moving
```

```python
# 長野県野沢温泉村
df_nagano = pd.read_excel("../data/2023_長野県_移動状況.xlsx", header=None)
df_nozawaonsen = df_nagano[df_nagano[1] == "野沢温泉村"]
obuse_jan = df_obuse.iloc[0, [2, 3, 4, 5]].values
obuse_dec = df_obuse.iloc[0, [46, 47, 48, 49]].values
difference = obuse_dec - obuse_jan

header = ["世帯数", "総計", "男", "女"]
df_nozawaonsen_moving = pd.DataFrame([difference], columns=header)
df_nozawaonsen_moving
```

```python
# 秋田県大潟村
df_ogata_moving = pd.read_excel("../data/2023_秋田県_移動状況.xlsx", sheet_name="Sheet1")

over_65 = ("65", "70", "75", "80", "85", "90", "95", "100")
over_65_sum = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["総計"].sum()
over_65_sum_man = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["男"].sum()
over_65_sum_woman = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["女"].sum()
over_65_in_inner_man = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["転入_県内_男"].sum()
over_65_in_inner_woman = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["転入_県内_女"].sum()
over_65_in_outer_man = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["転入_県外_男"].sum()
over_65_in_outer_woman = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["転入_県外_女"].sum()
over_65_out_inner_man = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["転出_県内_男"].sum()
over_65_out_inner_woman = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["転出_県内_女"].sum()
over_65_out_outer_man = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["転出_県外_男"].sum()
over_65_out_outer_woman = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["転出_県外_女"].sum()
over_65_dead_man = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["死亡_男"].sum()
over_65_dead_woman = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["死亡_女"].sum()
over_65_different_man = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["差引増減_男"].sum()
over_65_different_woman = df_ogata_moving[df_ogata_moving["Unnamed: 0"].str.startswith(over_65)]["差引増減_女"].sum()

df_ogata_moving["Unnamed: 0"] = df_ogata_moving["Unnamed: 0"].apply(lambda x: f"{x}歳")
df_ogata_moving = pd.concat([df_ogata_moving[0:13], pd.DataFrame(
    {
        "Unnamed: 0": "65歳以上", 
        "総計": over_65_sum,
        "男": over_65_sum_man,
        "女": over_65_sum_woman,
        "転入_県内_男": over_65_in_inner_man,
        "転入_県内_女": over_65_in_inner_woman,
        "転入_県外_男": over_65_in_outer_man,
        "転入_県外_女": over_65_in_outer_woman,
        "転出_県内_男": over_65_out_inner_man,
        "転出_県内_女": over_65_out_inner_woman,
        "転出_県外_男": over_65_out_outer_man,
        "転出_県外_女": over_65_out_outer_woman,
        "死亡_男": over_65_dead_man,
        "死亡_女": over_65_dead_woman,
        "差引増減_男": over_65_different_man,
        "差引増減_女": over_65_different_woman
    }, index=[0])], ignore_index=True)
df_ogata_moving.rename(columns={"Unnamed: 0": "5歳階級別"}, inplace=True)
df_ogata_moving.tail()
```
